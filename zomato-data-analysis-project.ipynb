**ZOMATO COMPANY DATA ANALYSIS PROJECT :**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# create the dataframe
df= pd.read_csv("/kaggle/input/zomato-dataset/Zomato data .csv")
df.head()
```

```python
df
```

**DATA CLEANING :**

```python
def HandleRate(value):

    value= str(value).split('/')
    value=value[0];
    return float(value)

df["rate"]= df["rate"].apply(HandleRate)
df.head()
```

**SUMMARY OF DATASET :**

```python
df.info()
```

**TYPE OF RESTAURANT :**

```python
df.head()
```

```python
sns.countplot(x=df["listed_in(type)"])
plt.xlabel("Type Of Restaurant")
```

```python
# CONCLUSION:
# The majority of the customers falls inte the Dining Category,Dining Restaurants are preferred by a larger number of individuals.
```

**VOTES RECEIVED FROM CUSTOMERS :**




```python
df.head()
```

```python
grouped_data= df.groupby("listed_in(type)")["votes"].sum()
result= pd.DataFrame({'votes': grouped_data})
plt.plot(result, c='green', marker="o")
plt.xlabel("Type of Restaurant",c="red", size=20)
plt.ylabel("Votes",c="red",size=20)
plt.show()
```

```python
# CONCLUSION :
# Dining restaurants has received maximum Votes##
```

**MAJORITY RESTAURANTS RECEIVED RATINGS :**

**WHICH MODE(ONLINE/OFFLINE) HAS RECEIVED THE MAXIMUM RATINGS :**

```python
df.head()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x=df["online_order"],y=df["rate"])
```

```python
# CONCLUSION :
# ONLINE ORDERS RECEIVE HIGHER RATINGS THEN OFFLINE ORDERS
```

**WHICH TYPE OF RESTAURANTS RECEIVE MORE OFFLINE ORDERS ? :**

```python
df.head()
```

```python
pivot_table= df.pivot_table(index="listed_in(type)", columns="online_order", aggfunc="size", fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("HeatMap")
plt.xlabel("Online Order")
plt.ylabel("listed_in(type)")
plt.show()
```

```python
# CONCLUSION:
# DINING RESTAURANTS PRIMARILY RECEIVE OFFLINE ORDERS, WHEREAS CAFES PRIMARILY RECEIVE ONLINE ORDERS,THIS SUGGESTS THAT CLIENTS PREFER TO PLACE ORDERS IN PERSON AT RESTAURANTS,BUT PREFER ONLINE ORDERING AT CAFES
```
